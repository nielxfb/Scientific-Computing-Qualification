```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
# 50 Random points range 50-100
x = np.array(np.sort(np.random.uniform(50, 100, 50)))
# 50 Random points range 0-20
y = np.array(np.sort(np.random.uniform(0, 20, 50)))

print(f"X = {x}")
print(f"Y = {y}")
```

```python
# Direct Inverse
A = np.vstack([x, np.ones(len(x))]).T
y_axis = y[:, np.newaxis]

alpha = np.dot(np.linalg.inv(np.dot(A.T, A)), np.dot(A.T, y_axis))
m = alpha[0]
c = alpha[1]

print(f"Direct Inverse - m = {m}, c = {c}")
```

```python
# Pseudoinverse
A = np.vstack([x, np.ones(len(x))]).T
y_axis = y[:, np.newaxis]

alpha = np.dot(np.linalg.pinv(A), y_axis)
m = alpha[0]
c = alpha[1]

print(f"Pseudoinverse - m = {m}, c = {c}")
```

```python
figure(figsize=(8, 6), dpi=80)
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x, y, 'b.', label='x and y coordinates')
plt.plot(x, m*x + c, 'r', label='Regression Line')
plt.legend()
plt.show()
```
